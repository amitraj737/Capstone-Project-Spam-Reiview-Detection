```python
# Created a function that finds jaccard distance based on the common bi-gram to the total number of bi-gram.
def jaccard_distance(a, b):
    """Calculate the jaccard distance between sets A and B"""
    a = set(a)
    b = set(b)
    if (len(a|b)!=0):
        return float(len(a&b))/len(a|b)
    else:
        return(0)
```

```python
def Two_Gram(i):                                      #Function definition for converting the data into a bi-gram
    bigrm1 = list(nltk.bigrams(str(i).lower().split())) #used nltk package to convert in bigrams 
    bigrm1 = list(map(lambda x:str(x[0]+' '+x[1]),bigrm1))   #firstly converting into lower case and then followed by splitting the data
    return(bigrm1)                                           #return bigram as a list.



for i in df_test.reviewText:
    compare_list.append(Two_Gram(i))      
```

```python
df_test['bigram'] = compare_list       #creating a new column named bigram.
```

```python
#Created a function to to calculate similarity within each reviewerID

def similarity_within_reviewer(df):
    for i in range(0,len(df)):
        for j in range(i+1,len(df)):
            val = jaccard_distance(df.bigram[i],df.bigram[j])
            if(val >0.9):
                value.append(val)
                from_no.append(df.row_num[i])
                to_no.append(df.row_num[j])
```

```python
similarity_within_reviewer(df_test)  #Faunction call for 'similarity_within_reviewer' passing dataframe a parameter.
```

```python
df_score['reviewer_simiValue'] = value    #creating new column containing  similarity score value
df_score['reviewer_from_no'] = from_no    #creating new column containing reviewerID from  
df_score['reviewer_to_no'] = to_no        #creating new column containing reviewerID to
```
